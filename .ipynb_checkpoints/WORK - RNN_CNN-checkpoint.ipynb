```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure


path = 'DataSet/'
state = 'Full_Data'
full_path = path+state+'.csv'
data_set = pd.read_csv(full_path)
criteria = "Occurence"

def get_list_states(data_set_ = data_set):
    return data_set["State"].drop_duplicates()

def get_list_names(length_=-1, data_set_ = data_set):
    return data_set_.pivot_table(index = "Year", columns="Name", values=criteria).sum(axis=0).sort_values(ascending=False).index[:length_]
```

#### We select the names we will work on

```python
input_size = 100
name_list = get_list_names(input_size)
name_list.shape
```

#### We select the states we will work on

```python
state_list = get_list_states()
state_list.shape
```

#### We Select and Normalize the related Data

```python
numpy_data_set = data_set[data_set["State"].isin(state_list.tolist())][data_set["Name"].isin(name_list.tolist())].pivot_table(index = ["State","Year"], columns=["Name"], values="Occurence").fillna(0).values.reshape(len(state_list), -1, len(name_list))

# normalized_data_set = numpy_data_set/np.sum(numpy_data_set, axis=1)[:, None, :]

# train_input_data = data_set[data_set["State"]=="CA"][data_set["Name"].isin(name_list.tolist())].pivot_table(index = "Year", columns="Name", values="Occurence").fillna(0).values
# test_input_data = data_set[data_set["Name"].isin(name_list.tolist())].pivot_table(index = "Year", columns="Name", values="Occurence").fillna(0).values

# train_epoch = np.array([normalized_data_set[i] for i in range(len(normalized_data_set-1))])
# test_epoch  = [normalized_data_set[i] for i in range(len(normalized_data_set-1))]

print("numpy_data_set.shape = ", numpy_data_set.shape)
```

###### Normalize so that we have the % of occurence for any year
> sum(data, axis = 0) → Sum(State| Year,  Name) <br/>
> sum(data, axis = 1) → Sum(Year | State, Name) <br/>
> sum(data, axis = 2) → Sum(Name | State, Year) <br/>

```python
axis = 2
normalized_data_set = numpy_data_set/np.sum(numpy_data_set, axis=axis)[:, :, None]
```

## Epoch Generation

```python
test_state_id = 4


train_epoch = np.delete(normalized_data_set, test_state_id, axis=0)
test_epoch = normalized_data_set[test_state_id].reshape(1, normalized_data_set.shape[1], normalized_data_set.shape[2])

print("train_epoch.shape = ",train_epoch.shape)
print("test_epoch.shape  = ",test_epoch.shape)
```

## Functions 
<em> To select precise batchs</em>
1. To select batchs (since we only have 51) <br/>
2. To select random values within this batch(since we only have 108) <br/>
> \>> get_random_batch : Selects randomly a batch from an epoch <br/>
> \>> get_splitted_batch : Creates a batch by splitting another batch into windows <br/>
> \>> get_splitted_batch : Creates a batch by taking random values from another batch <br/>

```python
def get_random_batch(epoch):
    return epoch[np.random.randint(epoch.shape[0])]

#TO FINISH ?
def get_splitted_batch(batch, window_size, prediction_size):
    temp_batch_size = 108 - window_size - prediction_size
    return np.array([batch[i: i-prediction_size-temp_batch_size] for i in range(temp_batch_size)]) ,\
            np.array([batch[i-prediction_size-temp_batch_size: i-temp_batch_size] for i in range(temp_batch_size)])

def get_randomized_batch(batch, batch_size, window_size, prediction_size):
    random_indexes = [np.random.randint(batch.shape[0]-window_size-prediction_size) for i in range(batch_size)]
    return np.array([batch[random_index:random_index+window_size] for random_index in random_indexes]) ,\
            np.array([batch[random_index+window_size:random_index+window_size+prediction_size] for random_index in random_indexes])
```

# TENSORFLOW PART

## TF Model

```python
#tensorboard --logdir C:\Users\louis\Desktop\Tsinghua\Machine_Learning\ML_Project\logs

prediction_size = 10
window_size = 50
batch_size = 200

import tensorflow as tf
tf.reset_default_graph()

learning_rate = 0.001

# output_size = input_size
output_size = input_size

# CNN Variables
number_filter = 1.
strides = 1
kernel_size = 3

#RNN Variables
number_rnn_layers = 100


# GRAPH PART
with tf.name_scope('Inputs'):
    x = tf.placeholder(tf.float32, [None, window_size    ,input_size ], name="x_placeholder")   
    y = tf.placeholder(tf.float32, [None, prediction_size, output_size], name="y_placeholder")


with tf.name_scope('CNN'):
    cnn_output =  tf.layers.conv1d(inputs = x, filters = number_filter, strides = strides, kernel_size = kernel_size, padding = "SAME", name = "CNN")
#     cnn_output_reduced = tf.layers.max_pooling1d(inputs=cnn_output, pool_size=2, strides=2)


with tf.name_scope('RNN'):
    rnn_cell = tf.nn.rnn_cell.LSTMCell(number_rnn_layers)
    rnn_output, first_states = tf.nn.dynamic_rnn(rnn_cell, cnn_output, dtype=tf.float32) 

    
#TODO :  Check this part (stacked_outputs especially)
with tf.name_scope('Dense_Layer'):
    stacked_rnn_output = tf.reshape(rnn_output, [-1, window_size*number_rnn_layers]) 
    stacked_outputs = tf.layers.dense(stacked_rnn_output, prediction_size*output_size)   
    output = tf.reshape(stacked_outputs, [-1, prediction_size, output_size])   

# OPTIMIZATION PART
with tf.name_scope('Analysis'):
    loss = tf.reduce_sum(tf.square(output - y))/(tf.to_float(tf.shape(output)[0]))
    optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
    train = optimizer.minimize(loss)  
    
    
with tf.Session() as sess: 
    writer = tf.summary.FileWriter("logs/", sess.graph)
```

## TF Simulation

```python
%%time
with tf.Session() as sess:
    init = tf.global_variables_initializer()
    sess.run(init)
    

    epoch_size = 1000
    
#     train_batch_xs = train_input_batch
#     train_batch_ys = train_output_batch
 
#     test_batch_xs = test_input_batch
#     test_batch_ys = test_output_batch
 
    test_batch_xs, test_batch_ys = get_splitted_batch(test_epoch[0], window_size, prediction_size)
    
    # >> Training part
    for i_batch in range(epoch_size):
        
        raw_batch = get_random_batch(train_epoch)
        train_batch_xs, train_batch_ys = get_randomized_batch(raw_batch, batch_size, window_size, prediction_size)
        
        if i_batch%100==0:
            print("")
            print("Iteration ", i_batch, " - Train Loss = ", sess.run(loss, feed_dict={x:train_batch_xs, y:train_batch_ys}))
            print(" >> Test Loss  = ", sess.run(loss, feed_dict={x:test_batch_xs, y:test_batch_ys}))
    
        
        sess.run(train, feed_dict={
            x: train_batch_xs,
            y: train_batch_ys,
        })    

    # << Training part
    
    # >> Generation part
    prediction = sess.run(output, feed_dict={
        x: test_batch_xs,
        y: test_batch_ys,
    })
    
#     name_id = 0
#     batch_id = 0
#     first_year = 1910+batch_id
#     plt.plot(range(first_year, first_year+window_size), test_batch_xs[batch_id,:,name_id],   label = "real_input")
#     plt.plot(range(first_year+window_size, first_year+window_size+prediction_size), test_batch_ys[batch_id,:,name_id] ,  label = "real_output")
#     plt.plot(range(first_year+window_size, first_year+window_size+prediction_size), prediction[batch_id,:,name_id], label = "pred_output")
#     plt.legend()
#     plt.plot()
```

```python
import sklearn
from sklearn.preprocessing import normalize

n_inputs = nbr_states
n_classes = nbr_states

min_val = 4

def get_xy(list_values_, n_steps_):
    x_ = np.empty((len(list_values_) - n_steps_-1, n_steps_, n_inputs), float)
    y_ = np.empty((len(list_values_) - n_steps_-1, n_steps_, n_classes), float)
    
    for i in range(len(list_values_) - n_steps_-1):
        x_[i] = list_values[i:i+n_steps_]
        y_[i] = np.transpose(np.array(list_values[i+1:i+n_steps_+1]))[0:n_classes].reshape(50, n_classes)

    return x_, y_
        
    
# list_values = np.array([[j for i in range(n_inputs)] for j in range(length_data)])
list_values = np.array(get_year_list("Mary").reindex(range(1910,2017)).fillna(min_val))
list_values = list_values/np.max(list_values, axis=0) #Normalisation
x_values, y_values = get_xy(list_values, n_steps)
```

## TF

```python
import tensorflow as tf
tf.reset_default_graph()
r_neuron = 100    
n_windows=50
n_input=n_inputs
n_output=n_classes
n_hidden_layers=100


## 1. Construct the tensors

with tf.name_scope('inputs'):
    x = tf.placeholder(tf.float32, [1, n_windows, n_input], name="x")   
    y = tf.placeholder(tf.float32, [1, n_windows, n_output], name="y")

    
with tf.name_scope('Weights'):
    weights = {
        'first': tf.Variable(tf.truncated_normal([n_inputs, n_hidden_layers],
                                        stddev=1./10)), 
        'second': tf.Variable(tf.truncated_normal([n_inputs, n_hidden_layers],
                                        stddev=1./10))
    }
    
    
with tf.name_scope('Biases'):
    biases = {
        'first': tf.Variable(tf.constant(0.1, shape=[n_hidden_layers, ])),
        'second': tf.Variable(tf.constant(0.1, shape=[n_hidden_layers, ]))
    }

def trans_x(X, name):
    with tf.name_scope('Wx_plus_b'):
        X = tf.reshape(x, [-1, n_inputs])
        X_first = tf.matmul(X, weights[name]) + biases[name]
        X_first = tf.reshape(X_first, [1, n_windows, n_hidden_layers])
    return X_first


## 2. create the model
with tf.name_scope('cnn'):
    cnn_result =  tf.nn.conv1d(value = x, filters = 3., stride = 1, padding = "same", name = "CNN")
    

## 2. create the model
with tf.name_scope('first_cell'):
    first_cell = tf.nn.rnn_cell.LSTMCell(n_hidden_layers)
    
#cell = tf.contrib.rnn.MultiRNNCell([first_cell] * 3, state_is_tuple=True)
with tf.name_scope('RNN'):
    first_rnn_output, first_states = tf.nn.dynamic_rnn(first_cell, trans_x(x, "first"), dtype=tf.float32) 


with tf.name_scope('dense'):
    stacked_rnn_output = tf.reshape(first_rnn_output, [-1, n_hidden_layers])          
    stacked_outputs = tf.layers.dense(stacked_rnn_output, n_output)       
    outputs = tf.reshape(stacked_outputs, [-1, n_windows, n_output])   
      
# stacked_outputs = tf.layers.dense(first_rnn_output, n_output)     

## 3. Loss + optimization
learning_rate = 0.01  
with tf.name_scope('loss'):
    loss = tf.reduce_sum(tf.square(outputs - y))    
    
    
with tf.name_scope('optimizer'):
    optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate) 

with tf.name_scope('train'):
    training_op = optimizer.minimize(loss)   
with tf.Session() as sess: 
    writer = tf.summary.FileWriter("logs/", sess.graph)
```
